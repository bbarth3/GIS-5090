```python
fc = "Airports.shp"
unique_name = arcpy.CreateUniqueName("airports_buffer")
delim_field = arcpy.AddFieldDelimiters(fc, "Fac_Type")
```

```python
sql_exp = delim_field + " = 'AIRPORT'"
arcpy.Select_analysis(fc, unique_name, sql_exp)
arcpy.Buffer_analysis(fc, unique_name, "1 MILE")
```

```python
inFeatures = ["Populated_Places.shp", "airports_buffer"]
intersectOutput = "places_by_airports"
arcpy.Intersect_analysis(inFeatures, intersectOutput, "", "", "point")
```

```python
arcpy.AddField_management("places_by_airports", "POP_Status", "TEXT")
```

```python
arcpy.AddField_management("places_by_airports", "POP_Percent", "LONG")
```

```python
fc = r'C:\Users\brend_000\Documents\ArcGIS\Projects\GIS 5090\GIS 5090 - Final Project\GIS 5090 - Final Project.gdb\places_by_airports'
fields = ['POP_2000', 'POP_2010', "POP_Status", "POP_Percent", "OBJECTID_12"]

# Create update cursor for feature class 
with arcpy.da.UpdateCursor(fc, fields) as cursor:
    for row in cursor:
        status = ((row[1] - row[0])/row[0])*100
        row[3] = status
        if status < 0:
            row[2] = "Declining"
        elif (status >= 0 and status < 3):
            row[2] = "Stable"
        else:
            row[2] = "Growing"
        print(str(row[4]) + " " + str(status) + " " + str(row[2]))
        cursor.updateRow(row)
```
